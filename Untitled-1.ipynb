```python
import pandas as pd
from geopy.geocoders import Nominatim

# Read the CSV file
csv_file = 'coordenadas_v2.csv'
df = pd.read_csv(csv_file)

# Initialize the geocoder
geolocator = Nominatim(user_agent="geoapiExercises")

# Function to get address from latitude and longitude
def get_address(row):
    try:
        location = geolocator.reverse((row['Latitude'], row['Longitude']), exactly_one=True)
        return location.address if location else None
    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
# Apply the get_address function to each row in the DataFrame
df['Address'] = df.apply(get_address, axis=1)

# Save the updated DataFrame to a new CSV file
output_csv_file = 'locations_with_addresses.csv'
df.to_csv(output_csv_file, index=False)

print(f"Addresses have been added to '{output_csv_file}'.")
```

```python
from geopy import distance
import pandas as pd
from datetime import datetime
import time

startTime = datetime.now()
print(datetime.now() - startTime)
lat1 = 18.4586700000
long1 = -65.9496523000

def distance_2points(row):
    coord1 = (lat1, long1)
    # convert the latitude and longitude to floats
    row['Latitude'] = float(row['Latitude'])
    row['Longitude'] = float(row['Longitude'])
    # create a tuple from the lat and long
    coord2 = (row['Latitude'], row['Longitude'])
    # return the distance in miles
    results = distance.distance(coord1, coord2).miles
    #print(f"Distance: {results}")
    return results

df = pd.read_csv('coordenadas_v2.csv')
df = df.head(12000)
df['distance'] = df.apply(lambda row: distance_2points(row), axis=1)
print(datetime.now() - startTime)

# get points within 5 miles
miles = 5
new_df = df[df['distance'] <= miles]
print(new_df)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import usaddress

# Step 2: Read the CSV file into a DataFrame
df = pd.read_csv('files/coordenadas_v3_0.csv')

# Step 3: Create a geocoder object (Nominatim is a free geocoding service)
geolocator = Nominatim(user_agent="reverse_geocoder")

# Step 4: Reverse geocode each pair of latitude and longitude
def reverse_geocode(row):
    location = geolocator.reverse((row['latitude'], row['longitude']))
    print(f"{row.name}: {location}")
    return location.address if location else None
  
def parse_address(row):
    address = usaddress.parse(row['Reverse_Geocoded_Address'])
    if address:
        return address[0][0]
    else:
        return None

df['Reverse_Geocoded_Address'] = df.apply(reverse_geocode, axis=1)

df['Parsed_Address'] = df.apply(parse_address, axis=1)

# Step 5: You now have reverse geocoded results in a new column 'Reverse_Geocoded_Address'

# Step 6: Save the updated DataFrame to a new CSV file
df.to_csv('files/coordenadas_v3_0.csv', index=False)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://findenergy.com/pr/'

# get the HTML content and create a DataFrame with the table data
html_content = requests.get(url).text
soup = BeautifulSoup(html_content, "html.parser")
table = soup.find_all("table", attrs={"class": "table sortable-table svelte-x63klk"})[3]
table_data = table.tbody.find_all("tr")  # contains 2 rows
table_head = table.thead.find_all("tr")  # contains 1 row

# create a list of column names
headings = []
for th in table_head[0].find_all("th"):
    # remove any newlines and extra spaces from left and right
    headings.append(th.text.replace('\n', ' ').strip())
    
# create a list of rows
data = []
for tr in table_data:
    t_row = {}
    # loop through each td tag and grab the text
    for td, th in zip(tr.find_all("td"), headings):
        t_row[th] = td.text.replace('\n', '').strip()
        # get the link from the first column
        if th == 'County':
            t_row['Link'] = td.find('a')['href']
    data.append(t_row)
    
# create a DataFrame from the list of dicts
df = pd.DataFrame(data)

df['Link'] = df['Link'].apply(lambda x: f"https://findenergy.com{x}")

new_df = df.copy()

new_df
```
